```python
%sql
CREATE SCHEMA IF NOT EXISTS optio_warehouse.silver;
CREATE SCHEMA IF NOT EXISTS optio_warehouse.gold;
```

```python
%sql
USE CATALOG optio_warehouse;
USE SCHEMA silver;
--- Table definitions (idempotent) --
CREATE TABLE IF NOT EXISTS optio_warehouse.silver.silver_active_locks_daily (
  dt STRING,
  height BIGINT,
  snapshot_ts TIMESTAMP,
  address STRING,
  unlock_date DATE,
  denom STRING,
  amount_uopt DECIMAL(38,0),
  lock_id STRING,
  source_run_id STRING
)
USING DELTA
PARTITIONED BY (dt);

CREATE TABLE IF NOT EXISTS optio_warehouse.silver.silver_chain_daily_state (
  dt STRING,
  height BIGINT,
  snapshot_ts TIMESTAMP,
  total_supply_uopt DECIMAL(38,0),
  total_locked_uopt DECIMAL(38,0),
  bonded_uopt DECIMAL(38,0),
  not_bonded_uopt DECIMAL(38,0),
  source_run_ids MAP<STRING, STRING>
)
USING DELTA
PARTITIONED BY (dt);
```

```python
%sql
USE CATALOG optio_warehouse;
USE SCHEMA gold;

CREATE TABLE IF NOT EXISTS optio_warehouse.gold.gold_supply_stake_lock_daily (
  dt STRING,
  total_supply_uopt DECIMAL(38,0),
  staked_uopt DECIMAL(38,0),
  locked_uopt DECIMAL(38,0),
  liquid_est_uopt DECIMAL(38,0),
  source_run_ids MAP<STRING, STRING>,
  computed_at TIMESTAMP
)
USING DELTA
PARTITIONED BY (dt);

CREATE TABLE IF NOT EXISTS optio_warehouse.gold.gold_unlock_buckets_daily (
  dt STRING,
  unlock_bucket STRING,
  locked_uopt DECIMAL(38,0),
  lock_count BIGINT,
  wallet_count BIGINT,
  computed_at TIMESTAMP
)
USING DELTA
PARTITIONED BY (dt);

CREATE TABLE IF NOT EXISTS optio_warehouse.gold.gold_locked_holder_distribution_daily (
  dt STRING,
  holding_bucket STRING,
  wallet_count BIGINT,
  total_locked_uopt DECIMAL(38,0),
  computed_at TIMESTAMP
)
USING DELTA
PARTITIONED BY (dt);

CREATE TABLE IF NOT EXISTS optio_warehouse.gold.gold_lock_calendar_daily (
  dt STRING,
  unlock_date DATE,
  days_to_unlock INT,
  unlocking_uopt DECIMAL(38,0),
  wallet_count BIGINT,
  lock_count BIGINT,
  computed_at TIMESTAMP
)
USING DELTA
PARTITIONED BY (dt);
```
